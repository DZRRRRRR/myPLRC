```python
import numpy as np
import torch
import os
from PLRC.datasets.DAPreDataset import DAPre
import cv2
from  matplotlib import pyplot as plt
```

```python
DS = DAPre(0,None,"../dataset/OfficeHomeDataset/")
errorlist = np.load("./errorlist.npy",allow_pickle=True)
```

```python
set = []
for d in errorlist:
    dStr = d.keys().__str__()
    set.append(int(dStr.split('[')[-1].split(']')[0]))
    
for i in set:
    print(i)
```

```python
for d in errorlist:
    dStr = d.keys().__str__()
    spdStr = dStr.split('[')[-1].split(']')[0].split(',')
    # print(spdStr)
    for s in spdStr:
        path = DS._imdb[int(s)]["im_path"]
        img = cv2.imread(path,1)
        im2 = img[:,:,::-1]
        plt.imshow(im2)
        plt.show()
```
